```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
import data

dataset = data.raw_chars()
print("Loaded {} characters".format(len(dataset)))
```

```python
ROWS = 10
COLS = 10
sample = random.sample(dataset, ROWS*COLS)
```

```python
plt.figure(figsize=(10, 14))

for i, s in enumerate(sample):
    plt.subplot(ROWS, COLS, i + 1)
    plt.axis('off')
    plt.title(s.character_id)
    for stroke in s.strokes:
        x, y = stroke.T
        plt.plot(x, -y)
plt.show()
```
